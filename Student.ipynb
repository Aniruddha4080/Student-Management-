```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(20,10))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=0.5, hspace=0.2)

plt.subplot(141)
plt.title('GENDER', fontsize = 20)
st['gender'].value_counts().plot.pie(autopct="%1.1f%%")

plt.subplot(142)
plt.title('RACE', fontsize = 20)
st['race/ethnicity'].value_counts().plot.pie(autopct="%1.1f%%")                           

plt.subplot(143)
plt.title('PARENTAL LEVEL OF EDUCATION', fontsize = 20)
st['parental level of education'].value_counts().plot.pie(autopct="%1.1f%%")  

plt.subplot(144)
plt.title('LUNCH', fontsize = 20)
st['lunch'].value_counts().plot.pie(autopct="%1.1f%%")  
```

 Let's have a looks on scores : 

```python
plt.rcParams['figure.figsize']=(20,10)
sns.countplot(st['reading score'], palette = 'RdPu', saturation=1, edgecolor=(0,0,0), linewidth=1 )
plt.title('READING SCORE', fontsize = 20)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(20,10)
sns.countplot(st['math score'], palette = 'RdBu', saturation=1, edgecolor=(0,0,0), linewidth=1 )
plt.title('MATH SCORE', fontsize = 20)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(20,10)
sns.countplot(st['writing score'], palette = 'pastel', saturation=1, edgecolor=(0,0,0), linewidth=1 )
plt.title('WRITING SCORE', fontsize = 20)
plt.show()
```

```python
# Distribution of marks
```

```python
plt.figure(figsize=(10,4))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=0.5, hspace=0.2)

plt.subplot(121)
plt.title('MATH SCORE')
sns.violinplot(y='math score',data=st)

plt.subplot(122)
plt.title('READING SCORE')
sns.violinplot(y='reading score',data=st)

plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9, wspace=0.5, hspace=0.2)

plt.subplot(121)
plt.title('WRITING SCORE')
sns.violinplot(y='writing score',data=st)
plt.show()
```

```python
 # lets compare score of both the gender
    # In maths male has score the most
    # In reading and writing female has score the most
```

```python
plt.rcParams['figure.figsize']=(12,5)
plt.subplot(131)
plt.title('MATH SCORE')
sns.barplot(x="gender",y="math score",data=st)

plt.subplot(132)
plt.title('READING SCORE')
sns.barplot(x="gender",y="reading score",data=st)

plt.subplot(133)
plt.title('WRITING SCORE')
sns.barplot(x="gender",y="writing score",data=st)
plt.show()
```

```python
# Comparing race/ethnicity with score
# Group E as scored most of the marks 
# Group A as scored less marks
```

```python
plt.rcParams['figure.figsize']=(15,8)
plt.subplot(131)
plt.title('MATH SCORE')
sns.barplot(x="race/ethnicity",y="math score",data=st)

plt.subplot(132)
plt.title('READING SCORE')
sns.barplot(x="race/ethnicity",y="reading score",data=st)

plt.subplot(133)
plt.title('WRITING SCORE')
sns.barplot(x="race/ethnicity",y="writing score",data=st)
plt.show()
```
